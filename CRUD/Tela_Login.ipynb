```python
#campos de texto

#declare esta variável para usar na consulta de campos em branco
flag=''

#df = pd.DataFrame()

p_nome = pn.widgets.TextInput(
    name = "Primeiro Nome",
    value='',
    placeholder='Digite o nome',
    disabled=False
)

s_nome = pn.widgets.TextInput(
    name = "Segundo Nome",
    value='',
    placeholder='Digite o nome',
    disabled=False
)

uk_email = pn.widgets.TextInput(
    name = "E-mail",
    value='',
    placeholder='Digite seu e-mail',
    disabled=False
)

senha = pn.widgets.TextInput(
    name="senha ",
    value='',
    placeholder='Digite sua senha',
    disabled=False
)

uk_matricula = pn.widgets.TextInput(
    name="matricula ",
    value='',
    placeholder='Digite sua matricula',
    disabled=False
)

link_telegram = pn.widgets.TextInput(
    name=" Link Telegram ",
    value='',
    placeholder='Digite o link do Telegram',
    disabled=False
)

datanasc = pn.widgets.DatePicker(
    name='Data de Nascimento',
    disabled=False
)

usr_role = pn.widgets.RadioBoxGroup(
    name='Tipo de usuário', options=['ADMIN', 'ALUNO', 'PROFESSOR'])

salario = pn.widgets.FloatInput(name="Salario", value=1000., step=1e-1, start=0, end=10000000)

buttonConsultar = pn.widgets.Button(name='Consultar', button_type='default')

buttonInserir = pn.widgets.Button(name='Inserir', button_type='default')

buttonExcluir = pn.widgets.Button(name='Excluir', button_type='default')



def queryAll():
    query = f"select * from usuario"
    df = pd.read_sql_query(query, cnx)
    return pn.widgets.Tabulator(df)

# consultar        
# neste exemplo o método de consulta usa o dataframe do pandas como retorno. Note que a flag é usada para ignorar quando um 
# campo for null (condição é sempre verdadeira). Veja que para cpf, que é uma string, foi usado '{cpf.value})' como parametro
# e para dnr que é numérico, foi usado {dnr.value} (sem aspas simples).
def on_consultar():
    try:  
        query = f"select * from usuario where ('{p_nome.value_input}'='{flag}' or p_nome='{p_nome.value_input}')"
        df = pd.read_sql_query(query, cnx)
        table = pn.widgets.Tabulator(df)
        return table
    except:
        return pn.pane.Alert('Não foi possível consultar!')


def on_inserir():
    try:            
        cursor= con.cursor()
        cursor.execute("insert into usuario(usr_role,p_nome,s_nome,uk_email, senha,  uk_matricula, link_telegram) VALUES (%s, %s, %s, %s, %s, %s, %s)", 
                    (usr_role.value,p_nome.value_input,s_nome.value_input, uk_email.value_input, senha.value_input, uk_matricula.value_input,  link_telegram.value_input))
        cursor.query
        con.commit()
        return queryAll()
    except:
        cursor.execute("ROLLBACK")
        cursor.close()
        return pn.pane.Alert('Não foi possível inserir!')



def on_excluir():
    try:
        cursor= con.cursor()
        cursor.execute("delete from usuario where uk_matricula=%s", (uk_matricula.value,))
        rows_deleted = cursor.rowcount
        con.commit()
        return queryAll()
    except:
        cursor.execute("ROLLBACK")            
        cursor.close() 
        return pn.pane.Alert('Não foi possível excluir!')

def table_creator(cons, ins, exc):
    if cons:
        return on_consultar()
    if ins:
        return on_inserir()
    if exc:
        return on_excluir()
    

interactive_table = pn.bind(table_creator, buttonConsultar, buttonInserir, buttonExcluir)
```

```python
pn.Row(pn.Column('Login', p_nome,s_nome,uk_email, senha,uk_matricula, usr_role, link_telegram, 
            pn.Row( buttonConsultar),
            pn.Row( buttonInserir),
            pn.Row( buttonExcluir)),
            pn.Column(interactive_table)).servable()
```
